**Theory Question**


1. What is NumPy, and why is it widely used in Python?
    - NumPy is a core Python library for fast, efficient numerical computing, offering high-performance multi-dimensional arrays and a wide range of mathematical functions to handle large datasets. It is widely used in data science and scientific applications because it simplifies and speeds up array operations through vectorization, broadcasting, and seamless integration with other data tools.

2. How does broadcasting work in NumPy?
    - Broadcasting in NumPy lets arrays of different shapes be used together in arithmetic operations without explicit reshaping. When two arrays are operated on, NumPy automatically expands the smaller array along any dimensions where its size is 1 or missing, so that both arrays have compatible shapes for element-wise operations. This process is efficient, avoids unnecessary copying, and greatly simplifies code for mathematical and data manipulation tasks.

3. What is a Pandas DataFrame?
    - A Pandas DataFrame is a two-dimensional, labeled data structure used for organizing, analyzing, and manipulating data in rows and columns, similar to a spreadsheet or SQL table. Each column can store a different data type, and DataFrames support various operations like sorting, filtering, merging, and grouping.

4. Explain the use of the groupby() method in Pandas.
    - The groupby() method in Pandas splits the DataFrame into groups based on column values, applies functions (like aggregation or transformation) to each group, and then combines the results. It follows a split-apply-combine approach, making it vital for summarizing or transforming data by categories, such as finding mean sales by product.

5. Why is Seaborn preferred for statistical visualizations?
    - Seaborn is preferred for statistical visualizations because it offers concise syntax, attractive default themes, and built-in support for visualizing statistical relationships and distributions. It also integrates smoothly with Pandas DataFrames, providing tailored plotting functions for data exploration and presentation.

6. What are the differences between NumPy arrays and Python lists?
    - NumPy arrays store elements of the same data type in contiguous memory locations, making them more memory-efficient and faster for numerical computations than Python lists. NumPy supports fast, element-wise operations optimized in C, which makes it ideal for large-scale mathematical tasks.
    - Python lists are more flexible, allowing mixed data types and dynamic resizing, but they use more memory and are slower for numerical processing due to overhead and fragmented storage. Lists are best for general-purpose use, while NumPy arrays excel in performance-critical numeric computations.

7. What is a heatmap, and when should it be used?
    - A heatmap is a graphical representation where data values are displayed with color gradients to help identify patterns, trends, and anomalies within datasets. Use heatmaps for visualizing matrix-like data, correlation tables, time-series intensity changes, or to highlight hotspots in user behavior or system performance.

8. What does the term “vectorized operation” mean in NumPy?
    - A vectorized operation refers to the process of performing element-wise computations on entire arrays at once without explicit Python loops. Vectorization leverages highly optimized C-based routines within NumPy, making mathematical operations on large datasets both fast and concise.

9.  How does Matplotlib differ from Plotly?
    - Matplotlib excels at static and highly customized plots, its interactivity is limited(except with plugins) while Plotly is best for creating interactive, web-friendly visualizations.

10. What is the significance of hierarchical indexing in Pandas?
    - Hierarchical indexing (MultiIndex) in Pandas allows multiple index levels on axes, enabling structured data grouping and more complex data manipulation. This supports advanced queries and operations on higher-dimensional data within a two-dimensional DataFrame structure.

11. What is the role of Seaborn’s pairplot() function?
    - Seaborn's pairplot() function visualizes pairwise relationships between multiple numeric variables in a dataset by creating a grid of scatter plots and histograms. The diagonal plots show the distribution of each variable, while the off-diagonal plots display scatterplots for each pair, helping identify correlations and patterns.

12. What is the purpose of the describe() function in Pandas?
    - The describe() function in Pandas generates a summary of descriptive statistics for DataFrame columns, including count, mean, standard deviation, minimum, maximum, and percentiles. It quickly provides insights into the distribution and key characteristics of numeric or categorical data.

13. Why is handling missing data important in Pandas?
    - Handling missing data is essential as unaddressed gaps (NaN or None) in datasets can lead to incorrect analyses and misleading results. Methods like dropna() and fillna() allow for clean and accurate data analysis by removing or imputing missing values.

14. What are the benefits of using Plotly for data visualization?
    - Plotly offers interactive, visually appealing plots with easy-to-use syntax. It supports zooming, hovering, and custom controls such as sliders, making data exploration intuitive. Additionally, Plotly supports multiple output formats and collaboration by sharing plots online, and it has broad compatibility with languages like Python, MATLAB, and R.

15. How does NumPy handle multidimensional arrays?
    - NumPy uses the ndarray object to efficiently represent multidimensional arrays with homogeneous data types. It stores the data in contiguous memory blocks and provides fast element-wise operations, slicing, broadcasting, and reshaping capabilities to work seamlessly with arrays of any number of dimensions.

16. What is the role of Bokeh in data visualization?
    - Bokeh is a Python library focused on interactive and web-ready visualizations. It enables creating rich, customizable plots that can be embedded in web applications, supporting real-time streaming and handling large datasets with interactive features like zooming, panning, and tooltips.

17. Explain the difference between apply() and map() in Pandas.
    - apply() is used to apply a function along an axis of the DataFrame or to a Series, capable of handling complex transformations on rows or columns. map() is a simpler operation mainly for substituting or mapping values in a Series using a dictionary or function, typically for element-wise transformations.

18. What are some advanced features of NumPy?
    - NumPy supports broadcasting, vectorized operations, advanced indexing, memory-efficient array manipulation, linear algebra, Fourier transforms, random sampling, and integration with C/C++ and Fortran code for high performance computation.

19. How does Pandas simplify time series analysis?
    - Pandas offers specialized data structures like DatetimeIndex for time series indexing, built-in date and time functionality, resampling, shifting, rolling windows, and handling missing data tailored for time-based operations and analysis.

20. What is the role of a pivot table in Pandas?
    - A pivot table in Pandas summarizes and reshapes data by aggregating values across multiple dimensions. It allows grouping data by one or more keys and applying aggregation functions, enabling quick analysis and exploration of multi-dimensional datasets.

21. Why is NumPy's array slicing faster than Python's list slicing?
    - NumPy array slicing is faster because it operates on contiguous memory blocks and returns a view of the original array without copying data, which makes it memory-efficient and quick. In contrast, Python list slicing involves creating a new list and copying elements, which is slower due to extra overhead and fragmented memory storage.

22. What are some common use cases for Seaborn?
    - Seaborn is commonly used for statistical data visualization including distribution plots, categorical plots, regression analysis, heatmaps, and pairwise relationship plots. It is widely used for exploratory data analysis (EDA) to reveal patterns, correlations, and summary statistics in datasets.

```python
# How do you create a 2D NumPy array and calculate the sum of each row?

import numpy as np
array_2d = np.array([[1, 2, 3, 4],
                     [5, 6, 7, 8]])
print(array_2d)

row_sums = array_2d.sum(axis=1)
print(row_sums)
```

```python
#  Write a Pandas script to find the mean of a specific column in a DataFrame.

import pandas as pd

data = {
    'A': [10, 20, 30, 40],
    'B': [15, 25, 35, 45],
}

df = pd.DataFrame(data)

mean_A = df['A'].mean()

print("Mean of column A:", mean_A)
```

```python
# Create a scatter plot using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create scatter plot
plt.scatter(x, y, color='blue', alpha=0.6)
plt.title('Scatter Plot using Matplotlib')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```

```python
#  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'A': [10, 20, 30, 40],
    'B': [12, 25, 29, 41],
    'C': [5, 15, 25, 35]
}
df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Generate a bar plot using Plotly.

import plotly.express as px
import pandas as pd

data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [23, 45, 56, 78]
}
df = pd.DataFrame(data)

fig = px.bar(df, x='Category', y='Values', title='Bar Plot using Plotly')
fig.show()
```

```python
#  Write a program to perform element-wise multiplication of two NumPy arrays.

import numpy as np

array1 = np.array([1, 2, 3])
array2 = np.array([4, 5, 6])

result = array1 * array2

print("Element-wise multiplication result:", result)
```

```python
# Create a line plot with multiple lines using Matplotlib.

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 1)
y1 = x
y2 = x ** 2
y3 = x ** 1.5

plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = x^2')
plt.plot(x, y3, label='y = x^1.5')

plt.title('Line Plot with Multiple Lines')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

data = {
    'Name': ['Harshali', 'Satasvi', 'Shreya', 'Mahima', 'Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Score': [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)

threshold = 80

filtered_df = df[df['Score'] > threshold]

print(filtered_df)
```

```python
# Create a histogram using Seaborn to visualize a distribution.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

sns.histplot(data, bins=30, kde=True)

plt.title('Histogram with KDE using Seaborn')
plt.xlabel('Data values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Perform matrix multiplication using NumPy.

import numpy as np

matrix1 = np.array([[1, 2],
                    [3, 4]])
matrix2 = np.array([[5, 6],
                    [7, 8]])

result = np.dot(matrix1, matrix2)

print("Matrix 1:")
print(matrix1)
print("Matrix 2:")
print(matrix2)
print("Matrix multiplication result:")
print(result)
```

```python
# Create a 3D scatter plot using Plotly.

import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    'X': np.random.rand(50),
    'Y': np.random.rand(50),
    'Z': np.random.rand(50),
    'Category': np.random.choice(['A', 'B'], size=50)
})

fig = px.scatter_3d(df, x='X', y='Y', z='Z', color='Category',
                    title='3D Scatter Plot using Plotly')

fig.show()
```
